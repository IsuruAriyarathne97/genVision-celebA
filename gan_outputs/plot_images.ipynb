```python
import os
import torch
import matplotlib.pyplot as plt
from torchvision.io import read_image
from torchvision.utils import make_grid

# Function to load and stack 100 sample images
def load_sample_images(folder, count=100):
    image_paths = sorted([os.path.join(folder, f) for f in os.listdir(folder) if f.endswith(".png")])[:count]
    images = [read_image(p) / 255.0 for p in image_paths]  # Normalize to [0,1]
    return torch.stack(images)

# Load 100 images
generated_imgs = load_sample_images("generated")
interpolated_imgs = load_sample_images("improved_generated")

# Plotting function
def plot_image_grid(images, title, nrow=10):
    grid = make_grid(images, nrow=nrow, normalize=True, padding=2)
    plt.figure(figsize=(12, 12))
    plt.imshow(grid.permute(1, 2, 0))
    plt.axis("off")
    plt.title(title)
    plt.show()

# Show grids
plot_image_grid(generated_imgs, "100 Sample Generated Images")
plot_image_grid(interpolated_imgs, "100 Latent Space Interpolated Images")
```
